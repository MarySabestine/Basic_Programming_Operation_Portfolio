```python
import tkinter as tk
from tkinter import messagebox
import requests


class WeatherApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Weather Application")
        self.root.geometry("400x300")


        # Create a label for the city input
        self.city_label = tk.Label(root, text="Enter City Name:", font=("Arial", 14))
        self.city_label.pack(pady=10)


        # Create an entry for the city name
        self.city_entry = tk.Entry(root, font=("Arial", 14))
        self.city_entry.pack(pady=10)


        # Create a button to get the weather
        self.get_weather_button = tk.Button(root, text="Get Weather", command=self.get_weather, font=("Arial", 14))
        self.get_weather_button.pack(pady=10)


        # Create a label to display the weather information
        self.weather_info = tk.Label(root, text="", font=("Arial", 12), wraplength=300)
        self.weather_info.pack(pady=20)


    def get_weather(self):
        city = self.city_entry.get()
        api_key = "b312f822127bc14fbe2e2970572d525b"  # Replace with your OpenWeatherMap API key
        url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"


        try:
            response = requests.get(url)
            data = response.json()


            if response.status_code == 200:
                # Extract relevant information
                temperature = data['main']['temp']
                weather_description = data['weather'][0]['description']
                humidity = data['main']['humidity']
                wind_speed = data['wind']['speed']


                # Display the weather information
                weather_report = (f"Temperature: {temperature}Â°C\n"
                                  f"Description: {weather_description}\n"
                                  f"Humidity: {humidity}%\n"
                                  f"Wind Speed: {wind_speed} m/s")
                self.weather_info.config(text=weather_report)
            else:
                messagebox.showerror("Error", "City not found. Please try again.")


        except Exception as e:
            messagebox.showerror("Error", str(e))


if __name__ == "__main__":
    root = tk.Tk()
    app = WeatherApp(root)
    root.mainloop()
```
