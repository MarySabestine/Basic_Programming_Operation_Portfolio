```python
from tkinter import *


# Create the main window
root = Tk()
root.title("Simple Calculator")
root.geometry("400x600")
root.resizable(0, 0)


# Global variable to store the expression
expression = ""


# Function to update the expression in the input field
def btn_click(item):
    global expression
    expression += str(item)
    input_text.set(expression)


# Function to clear the input field
def bt_clear():
    global expression
    expression = ""
    input_text.set("")


# Function to evaluate the expression and display the result
def bt_equal():
    global expression
    try:
        result = str(eval(expression))
        input_text.set(result)
        expression = ""
    except Exception as e:
        input_text.set("Error")
        expression = ""


# Input field to display the expression/result
input_text = StringVar()
input_frame = Frame(root)
input_frame.pack(pady=10)


input_field = Entry(input_frame, textvariable=input_text, font=('Arial', 24), bd=10, insertwidth=4, width=14, borderwidth=4)
input_field.grid(row=0, column=0)


# Buttons for calculator
button_frame = Frame(root)
button_frame.pack()


buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    'C', '0', '=', '+'
]


row_val = 0
col_val = 0


for button in buttons:
    if button == '=':
        Button(button_frame, text=button, padx=20, pady=20, font=('Arial', 18), command=bt_equal).grid(row=row_val, column=col_val)
    elif button == 'C':
        Button(button_frame, text=button, padx=20, pady=20, font=('Arial', 18), command=bt_clear).grid(row=row_val, column=col_val)
    else:
        Button(button_frame, text=button, padx=20, pady=20, font=('Arial', 18), command=lambda item=button: btn_click(item)).grid(row=row_val, column=col_val)


    col_val += 1
    if col_val > 3:
        col_val = 0
        row_val += 1


# Start the GUI event loop
root.mainloop()
```
